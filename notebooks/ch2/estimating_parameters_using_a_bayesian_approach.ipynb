Example: we want to estimate the probability of getting a head or a tail by a coin flip. In a frequentist setting, this problem is reduced to maximizing the likelihood:

$$
\mu_{ML} = \max_\mu P(\mu \mid x_1, \dots, x_N) = \max_\mu \prod_{i=1}^N \mu^{x_i}(1-\mu)^{1-x_i}
$$

From the notes, we have demonstrated that:

$$
\mu_{ML} = \frac1N \sum_{i=1}^{N} x_i = \frac{m}{N}
$$

Which is the sample mean (and $m$ is the number of heads given that heads is indicated by 1 and tail by 0). 

If we have 3 consecutive heads over 3 coin flips, then we will get $\mu=1$ (we will get head with 100% probability) which is unreasonable. This method is higly prone to overfitting.

We can use a Bayesian approach:

$$
p(\mu \mid D) = \frac{p(D \mid \mu) p(\mu)}{p(D)}
$$

The likelihood function measures the probability of getting $m$ heads out of $N$ coin flips, which (under the indipendence assumption) is modeled by the binomial distribution:

$$
p(D \mid \mu) = \binom{N}{m} \mu^m (1-\mu)^{(N-m)}
$$

Which prior do we use to model the parameter $\mu$? The beta distribution! The beta distribution is a distribution over a probability ($\mu$) and it's regulated by two hyperparameters $(a,b)$, where intuitively $a$ is proportional to the number of wins (or heads) and $b$ to the number of losts (or tails).

$$
p(\mu) = \frac{\Gamma(a+b)}{\Gamma(a) + \Gamma(b)} \mu^(a-1)(1-\mu)^{(b-1)}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma, binom

def beta(mu, a=3, b=2):
    """ beta distr. pdf """
    coeff = gamma(a + b) / (gamma(a) * gamma(b))
    value = (mu ** (a - 1)) * ((1 - mu) ** (b - 1))
    return coeff * value

def binomial(m, N, mu):
    coeff = binom(N, m)
    value = (mu ** m) * ((1 - mu) ** (N - m))
    return coeff * value
```

Let's see what happens for different values of (a,b)

```python
hyperparameters = [
    (0.1, 0.1), 
    (1.0, 1.0),
    (2.0, 3.0), 
    (8.0, 4.0)
]

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15, 15))
ax = ax.flatten()
X = np.linspace(0, 1, 500)
for i, h in enumerate(hyperparameters):
    Y = beta(X, *h)
    ax[i].plot(X, Y)
    ax[i].title.set_text(str(h))
    ax[i].set_xlim(left=0, right=1)
    ax[i].set_ylim(bottom=0, top=3)
plt.show()
```

Let's do an example:

1. flip the coin 150 times and create a dataset from the outcomes
2. Compute the dataset likelihood using the binomial distribution at different $\mu$
3. Set a=#heads, b=#tails, then compute the $\mu$ prior using the beta function
4. Compute the non-normalized posterior and observe the peak, that should be on 0.5 (which is the probability of getting head)

The posterior and the prior share the same functional relationship with $\mu$, which makes the prior conjugate to the posterior.

```python
# dataset
D = np.random.binomial(n=1, p=0.5, size=150)

# possible values for mu parameter
mus = np.linspace(0, 1, 500)
prior = beta(mus, a=D.sum(), b=(len(D) - D.sum()))
likelihood = binomial(m=D.sum(), N=len(D), mu=mus)
posterior = prior * likelihood # compute non-normalized posterior

fig, ax = plt.subplots(ncols=4, figsize=(16, 4))
ax[0].plot(mus, mus);        ax[0].title.set_text('mu value')
ax[1].plot(mus, prior);      ax[1].title.set_text('prior (beta)')
ax[2].plot(mus, likelihood); ax[2].title.set_text('likelihood (binomial)')
ax[3].plot(mus, posterior);  ax[3].title.set_text('posterior (non-normalized)')
plt.show()
```

We can express the dataset of coin flips as $m$ and $l$, that are the number of heads and tails respectively. Then if we multiply the prior and the likelihood and keep only the parameters dependent on $\mu$, we get:

$$
p(\mu \mid l, m, a, b) \propto \mu^{a + m - 1} (1-\mu)^{b + l - 1}
$$

Up to a normalization coefficient. We observe that the posterior is basically another beta function, and then by comparison with the function definition we find the normalization constant as:

$$
p(\mu \mid l, m, a, b) = \frac{\Gamma(a+m+b+l)}{\Gamma(a+m) + \Gamma(b+l)} \mu^{a + m - 1} (1-\mu)^{b + l - 1}
$$

We can see this approach as initializing $a,b$ with reasonable values, and then adjusting the posterior by observing the real data $l, m$. This approach can be also used in a sequential way, by increasing $l$ and $m$ when observing a value. 

In the following example we are going to initialize $a,b$ wrong, and then slowly add real coin flips to the posterior:

```python
l, m = 0, 0
init_a, init_b = 2, 1

mus = np.linspace(0, 1, 100)
iteration_sample_size = 2
iterations = 16
fig, ax = plt.subplots(nrows=4, ncols=4, figsize=(15, 15))
ax = ax.flatten()

posterior = beta(mus, a=init_a, b=init_b)
ax[0].plot(mus, posterior)
ax[0].title.set_text('initial state')

for i in range(1, iterations):
    sample = np.random.binomial(n=1, p=0.5, size=iteration_sample_size)
    l = iteration_sample_size - sample.sum()
    m = sample.sum()
    init_a += m
    init_b += l
    posterior = beta(mus, a=init_a, b=init_b)
    ax[i].plot(mus, posterior)
    ax[i].title.set_text(f'iteration {i}')

plt.show()
```

The posterior converges to the real posterior as the iterations increase. This sequential method is independent from the prior or likelihood we use and relies only of the i.i.d. assumption.

This sequential method works also in a frequentist setting. The following formula is obtained using the sum and product rules:

$$
p(x=1 \mid D) = \int_0^1 p(x=1 \mid \mu) p(\mu \mid D) d\mu = \int_0^1 \mu p(\mu \mid D) d\mu = \mathbb E[\mu \mid D]
$$

From the mean of the beta distribution we get:

$$
p(x=1 \mid D) = \frac{m + a}{m + a + l + b}
$$

which is the fraction of observed (m) and prior (a) wins against the number of tests (a+b+m+l). 

The bayesian and frequentist approaches agree for $N \to \infty$ (N is the number of samples / iterations)
